## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
# Вводим число
number = input("Введите пятизначное число: ")



result = number[0] + number[1:4][::-1] + number[4]

print(result)
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
# Вводим количество дней до отпуска (сразу превращаем в целое число — int)
days = int(input("Сколько дней осталось до отпуска? "))

# Считаем количество полных недель
weeks = days // 7

# В каждой полной неделе 2 выходных
weekends = weeks * 2

# Проверяем остаток дней (дни на последней неполной неделе)
remaining_days = days % 7

# Если остаток 6 дней, значит в него попала суббота (воскресенье было бы 7-м днем, но это уже полная неделя)
if remaining_days == 6:
    weekends += 1

print(weekends)
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
# Вводим данные
n = int(input("Введите длину плитки (n): "))
m = int(input("Введите ширину плитки (m): "))
k = int(input("Сколько долек нужно отломить (k): "))

# Проверяем условия:
# 1. k не больше всей площади
# 2. k делится на n БЕЗ остатка ИЛИ k делится на m БЕЗ остатка
if k <= n * m and (k % n == 0 or k % m == 0):
    print(True)
else:
    print(False)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
# Вводим число
num = int(input("Введите целое положительное число: "))

# Список кортежей: число и его римский эквивалент
# Важно идти от самых больших к самым маленьким
roman_numerals = [
    (1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'),
    (100, 'C'), (90, 'XC'), (50, 'L'), (40, 'XL'),
    (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')
]

result = ""

# Проходим по нашему списку
for value, symbol in roman_numerals:
    # Пока наше число больше или равно текущему значению из списка
    while num >= value:
        result += symbol  # Добавляем символ к ответу
        num -= value      # Вычитаем значение из числа

print(result)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
# Вводим данные
data = input("Введите число: ").strip()

# 1. Обработка отрицательных чисел (бонус)
# Если строка начинается с минуса и она длиннее одного символа, отрезаем минус
check_data = data
if check_data.startswith('-'):
    check_data = check_data[1:]

# 2. Проверка условий:
# - В строке не более одной точки: check_data.count('.') <= 1
# - Если удалить одну точку, остаются только цифры: check_data.replace('.', '', 1).isdigit()
# - Строка не пустая и не является просто точкой: check_data != "." and check_data != ""

if (check_data.count('.') <= 1 and 
    check_data.replace('.', '', 1).isdigit() and 
    check_data != "." and 
    check_data != ""):
    print(True)
else:
    print(False)
```
